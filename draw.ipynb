```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
alpha = 2

# Define the modified function to ensure symmetry around x=1
def modified_grad_x(x,factor=1):
    return alpha / 2 / (1 + (np.pi / 2 * alpha * (x - 1)/factor)**2)

# Generate x values
x = np.linspace(-3, 5, 400)

# Calculate y values using the modified function
y = modified_grad_x(x)

# Create the modified plot
plt.figure(figsize=(6, 5))
plt.plot(x, y, label=r'sg(x)')
plt.xlabel('x')
plt.ylabel('sg(x)')
plt.axvline(x=1, color='black', linestyle='--')
plt.xlim(-1, 3)
plt.legend()
plt.grid(True)
# plt.savefig("sgx.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
alpha = 2

# Define the modified function to ensure symmetry around x=1
def modified_grad_x(x, factor=1):
    return alpha / 2 / (1 + (np.pi / 2 * alpha * (x - 1) / factor)**2)

# Generate x values
x = np.linspace(-3, 5, 400)

# Factors to iterate over
factors = [1, 1.5, 2]

# Create the modified plot
plt.figure(figsize=(8, 6))
for factor in factors:
    y = modified_grad_x(x, factor)
    plt.plot(x, y, label=f'Factor = {factor}')

plt.xlabel('x')
plt.ylabel('sg(x)')
plt.axvline(x=1, color='black', linestyle='--')
plt.xlim(-0.5, 2.5)
plt.legend()
plt.grid(True)
plt.title('Modified sg(x) for Different Factors')
plt.savefig("modified_sgx_factors.pdf")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_sn_gradients(csv_file, sn):
    # CSV
    df = pd.read_csv(csv_file)
    
    # 
    columns = [
        f'sn{sn}_grad_1',
        f'sn{sn}_grad_before_2',
        f'sn{sn}_grad_before_3',
        f'sn{sn}_grad_before_4',
        f'sn{sn}_grad_after_2',
        f'sn{sn}_grad_after_3',
        f'sn{sn}_grad_after_4'
    ]
    
    
    # 
    data = df[columns]
    
    # 
    colors = ['black', 'blue', 'blue', 'blue', 'green', 'green', 'green']
    
    # 
    plt.figure(figsize=(10, 6))
    for i, col in enumerate(columns):
        plt.plot(data[col], label=col, color=colors[i])
    
    # 
    plt.legend()
    plt.title(f'sn{sn} Gradients')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()

# 
plot_sn_gradients('record_grad.csv', 16)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for the plot
x_values = np.linspace(-10, 10, 400)
# Compute the values for arctan(x)
arctan_x = np.arctan(x_values)
# Compute the values for 1.2 * arctan(x/1.2)
scaled_arctan_x = 1.2 * np.arctan(x_values / 1.2)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, arctan_x, label='arctan(x)', color='blue')
plt.plot(x_values, scaled_arctan_x, label='1.2 * arctan(x/1.2)', color='red')
plt.title('Comparison of arctan(x) and 1.2 * arctan(x/1.2)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  arctan 
def adjusted_arctan_x(x):
    return 0.5 + 2 * (np.arctan(x) / np.pi)

def adjusted_scaled_arctan_x(x, scale):
    return 0.5 + 2 * (scale * np.arctan(x / scale) / np.pi)

#  x  -1  1
x_values_focused = np.linspace(-1, 1, 400)

#  arctan  arctan 
adjusted_arctan_values_focused = adjusted_arctan_x(x_values_focused)
adjusted_scaled_arctan_values_focused = adjusted_scaled_arctan_x(x_values_focused, 2)

#  Heaviside 
heaviside_focused = np.heaviside(x_values_focused, 0.5)

# 
plt.figure(figsize=(8, 5))
plt.plot(x_values_focused, adjusted_arctan_values_focused, label='arctan(x)', color='orange')
plt.plot(x_values_focused, adjusted_scaled_arctan_values_focused, label='1.2 * arctan(x/1.2)', color='green')
plt.plot(x_values_focused, heaviside_focused, label='Heaviside function', color='blue', linestyle='--')
plt.title('Adjusted Arctan and Heaviside Functions (x in [-1, 1])')
plt.xlabel('x')
plt.ylabel('(x)')
plt.ylim(0, 1.01)
plt.xlim(-1, 1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  arctan 
def adjusted_arctan_x(x):
    return 0.5 + 2 * (np.arctan(x) / np.pi)

def adjusted_scaled_arctan_x(x, scale):
    return 0.5 + 2 * (scale * np.arctan(x / scale) / np.pi)

#  x  -1  1
x_values_focused = np.linspace(-1, 1, 400)

#  scale  arctan 
scaled_arctan_values = {}
scales = [1, 5, 9]
colors = ['orange', 'red', 'green'] 
# 'purple', 'blue', 'brown', 'pink', 'gray']

plt.figure(figsize=(12, 8))
for scale, color in zip(scales, colors):
    scaled_arctan_values[scale] = adjusted_scaled_arctan_x(x_values_focused, scale)
    plt.plot(x_values_focused, scaled_arctan_values[scale], label=f'{scale} * arctan(x/{scale})', color=color)

#  Heaviside 
heaviside_focused = np.heaviside(x_values_focused, 0.5)
plt.plot(x_values_focused, heaviside_focused, label='Heaviside function', color='black', linestyle='--')

plt.title('Adjusted Arctan and Heaviside Functions (x in [-1, 1])')
plt.xlabel('x')
plt.ylabel('(x)')
plt.ylim(0, 1.01)
plt.xlim(-1, 1)
plt.grid(True)
plt.legend()
plt.show()
```
